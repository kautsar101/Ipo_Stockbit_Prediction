```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
import my_email
import my_pass
import my_pin

# !!!! importance !!!!
# change this variable into you email, password and pin in stockbit
email = my_email.w
password = my_pass.w
pin = my_pin.w
```

```python
# change this path into your own path
path = 'C:/Users/Lenovo/Downloads/chromedriver/chromedriver.exe'
# Create a Service object
service = Service(executable_path=path)
# Initialize ChromeOptions
options = webdriver.ChromeOptions()
```

```python
# Create the WebDriver instance using the Service object
driver = webdriver.Chrome(service=service, options=options)
driver.maximize_window()
# Navigate to a website (for testing)
driver.get("https://stockbit.com/login")

# Perform any actions or tests you nee
email_field = driver.find_element(By.ID, "username")
password_field = driver.find_element(By.ID, "password")

# Quit the driver
email_field.send_keys(email)
password_field.send_keys(password)

# Find and click the login button
login_button = driver.find_element(By.ID, "email-login-button")
login_button.click()

skipbutton = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "modalnewavatar-button-skip")))
skipbutton.click()

# Locate the outermost "e-ipo" link by its href attribute and click it
e_ipo_link = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "a[href='/e-ipo']")))
e_ipo_link.click()

# Wait for the PIN modal to appear
pin_inputs = WebDriverWait(driver, 10).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "input[name^='auth-modal-input-pin']")))

# Enter the PIN
for i, digit in enumerate(pin):
    pin_inputs[i].send_keys(digit)

nanti_aja_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='ant-btn sc-13f62da7-5 sc-13f62da7-6 douxCm dPDhJX']/span[text()='Nanti Aja']")))
nanti_aja_button.click()

past_filter_item = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[text()='Past' and contains(@class, 'eipopagelist__FilterItem-jAqHTG')]")))
past_filter_item.click()
```

```python
stocks = driver.find_elements(By.CLASS_NAME, 'eipopagelist__EipoListItemContainer-bkfgwQ')
data = []
```

```python
# Extract details for each stock
stock_code = []
sectors = []
percentages = [] 
underwriters = []
names = []
dates = []
prices = []
subsectors = []
bidang_usahas = []
alamats = []
websites = []
jumlah_sahams = []
rasio_warrans = []
prospektus_links = []
summary_prospektus_links = []
image_sources = []
```

```python
image_elements = driver.find_elements(By.XPATH, "//img[@class='eipopagelist__EipoListItemEmitenLogo-ksJJJe znjzK']")
for image_element in image_elements:
    image_src = image_element.get_attribute('src')
    image_sources.append(image_src)
    print(f"image: {image_src}")
```

```python
from time import sleep
for stock in stocks:
    # Extract stock code
    stock_code_element = stock.find_element(By.CLASS_NAME, 'eipopagelist__EipoListItemEmitenCode-mVKAo')
    if 'eipopagelist__EipoListItemEmitenCode-mVKAo bpAEYg' in stock_code_element.get_attribute('class'):
        while True:
            # handle connection error while scraping
            try:
                print(f"CODE           : {stock_code_element.text}")
                stock_code.append(stock_code_element.text)

                # Extract company name
                company_name_element = stock.find_element(By.CLASS_NAME, 'eipopagelist__EipoListItemCompanyName-cBPlVP')
                company_name = company_name_element.text
                print(f"Company Name   : {company_name}")
                names.append(company_name)

                # Extract date
                date_element = stock.find_element(By.CLASS_NAME, 'eipopagelist__EipoListItemDate-iFkweb')
                date = date_element.text
                print(f"Date           : {date}")
                dates.append(date)

                # Extract price
                price_element = stock.find_element(By.CLASS_NAME, 'eipopagelist__EipoListItemPrice-cISyxC')
                price = price_element.text
                print(f"Price          : {price}")
                prices.append(price)
                
                # Click to navigate to the stock details page
                stock_code_element.click()

                # Extract sector
                sector_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Sektor']/following-sibling::p")))
                sector = sector_element.text
                print(f"Sector         : {sector}")
                sectors.append(sector)

                # Extract percentage
                persen_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Persenan Total Saham']/following-sibling::p")))
                percentage = persen_element.text
                print(f"Percentage     : {percentage}")
                percentages.append(percentage)

                # Extract underwriter
                admin_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Admin Partisipan']/following-sibling::p")))
                underwriter = admin_element.text
                print(f"Underwriter    : {underwriter}")
                underwriters.append(underwriter)

                # Extract subsektor
                subsektor_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Subsektor']/following-sibling::p")))
                subsektor = subsektor_element.text
                print(f"Subsector      : {subsektor}")
                subsectors.append(subsektor)

                # Extract bidang usaha
                bidang_usaha_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Bidang Usaha']/following-sibling::p")))
                bidang_usaha = bidang_usaha_element.text
                print(f"Bidang Usaha   : {bidang_usaha}")
                bidang_usahas.append(bidang_usaha)

                # Extract alamat
                alamat_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Alamat']/following-sibling::p")))
                alamat = alamat_element.text
                print(f"Alamat         : {alamat}")
                alamats.append(alamat)

                # Extract website
                website_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Website']/following-sibling::p")))
                website = website_element.text
                print(f"Website        : {website}")
                websites.append(website)

                # Extract jumlah saham yang ditawarkan
                jumlah_saham_element = WebDriverWait(driver, 20).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Jumlah Saham yang Ditawarkan']/following-sibling::p")))
                jumlah_saham = jumlah_saham_element.text
                print(f"Jumlah Saham   : {jumlah_saham}")
                jumlah_sahams.append(jumlah_saham)

                # Extract rasio warran
                try:
                    rasio_warran_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//h6[text()='Rasio Warran']/following-sibling::p")))
                    rasio_warran = rasio_warran_element.text
                except:
                    rasio_warran = 0
                print(f"Rasio Warran   : {rasio_warran}")
                rasio_warrans.append(rasio_warran)

                # Click "Prospektus" button and capture the URL
                prospektus_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[h6[text()='Prospektus']]")))
                prospektus_button.click()
                
                # Switch to the new tab
                driver.switch_to.window(driver.window_handles[1])
                prospektus_link = driver.current_url
                prospektus_links.append(prospektus_link)
                print(f"Prospektus Link: {prospektus_link}")

                # Close the new tab and switch back to the main tab
                driver.close()
                driver.switch_to.window(driver.window_handles[0])

                # Click "Ringkasan Prospektus" button and capture the URL
                ringkasan_button = WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//button[h6[text()='Ringkasan Prospektus']]")))
                ringkasan_button.click()
                
                # Switch to the new tab
                driver.switch_to.window(driver.window_handles[1])
                summary_prospektus_link = driver.current_url
                summary_prospektus_links.append(summary_prospektus_link)
                print(f"Summary Link   : {summary_prospektus_link}")

                # Close the new tab and switch back to the main tab
                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                print("-" * 80)
                break
            except:
                sleep(4)
                driver.refresh()
                continue
```

```python
import pandas as pd
data = {
    'Code': stock_code,
    'Name': names,
    'Date': dates,
    'Price': prices,
    'Share': percentages,
    'Amount': jumlah_sahams,
    'Warran': rasio_warrans,
    'Underwriter': underwriters,
    'Sector': sectors,
    'Subsector': subsectors,
    'Field': bidang_usahas,
    'Website': websites,
    'Prospectus': prospektus_links,
    'Summary_Prospectus': summary_prospektus_links,
    'Alamat': alamats,
    'Image':image_sources,
}

df = pd.DataFrame(data)
```
